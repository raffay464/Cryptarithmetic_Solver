## DFS ALGO

```python
def is_valid(puzzle, mapping):
    s1, s2, s3 = puzzle
    n1, n2, n3 = 0, 0, 0
    for c in s1:
        n1 = n1 * 10 + mapping[c]
    for c in s2:
        n2 = n2 * 10 + mapping[c]
    for c in s3:
        n3 = n3 * 10 + mapping[c]
    return n1 + n2 == n3

def solve(puzzle, mapping, index):
    if index == len(mapping):
        print("Mapping found:", mapping)
        return is_valid(puzzle, mapping)
    for i in range(10):
        if i in mapping.values():
            continue
        mapping[letters[index]] = i
        print("Trying:", mapping)
        if solve(puzzle, mapping, index + 1):
            return True
        print("Backtracking from:", mapping)
        mapping[letters[index]] = -1
    return False 
    

puzzle = ["SEND", "MORE", "MONEY"]
letters = list(set("".join(puzzle)))
mapping = {c: -1 for c in letters}
print(solve(puzzle, mapping, 0))

```

```python
def is_valid_assignment(assignment):
    """Check if the assignment satisfies the constraints."""
    # Ensure no leading zeros
    for key in assignment:
        if assignment[key] == 0 and key in unique_letters:
            return False
    
    # Check if the sum is valid
    sum_value = sum(assignment[key] for key in unique_letters)
    if sum_value != 0:
        return False
    
    return True

def solve_cryptarithmetic_dfs(puzzle, assignment={}, index=0):
    global unique_letters
    # Base case: if the assignment is complete, check if it's valid
    if len(assignment) == len(unique_letters):
        if is_valid_assignment(assignment):
            return assignment

    # Try assigning digits to letters recursively
    for digit in range(10):
        if digit in assignment.values():
            continue

        # Choose a letter to assign the current digit
        if index < len(unique_letters):
            current_letter = unique_letters[index]
            assignment[current_letter] = digit

            # Recursively try to solve with the new assignment
            result = solve_cryptarithmetic_dfs(puzzle, assignment, index + 1)
            if result:
                return result

            # Backtrack if the assignment is not valid
            del assignment[current_letter]

    return None

# Example usage
puzzle = "SEND + MORE = MONEY"
unique_letters = list(set(puzzle.replace(' ', '').replace('+', '').replace('=', '')))
solution = solve_cryptarithmetic_dfs(puzzle)
if solution:
    print("Solution found:")
    print(solution)
else:
    print("No solution found.")
```
